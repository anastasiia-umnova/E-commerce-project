Когорта — это группа пользователей, которые имеют общую характеристику, например, зарегистрировались в одном и том же месяце, совершили первую покупку в определенный период или использовали продукт в определённый временной промежуток.

```python
# месяц совершения первой покупки для каждого пользователя
ord_cus['month'] = ord_cus.order_approved_at.dt.to_period('M')
ord_cus['one_month'] = ord_cus.groupby('customer_unique_id')['order_approved_at'].transform('min').dt.to_period('M')
```

```python
# первые покупки за 2017 год
one_year = ord_cus.query('one_month >= "2017-01" and one_month <= "2017-12"')
```

```python
# разница между месяцем первого заказа и текущего
one_year['next_months'] = (one_year.month.dt.to_timestamp() - one_year.one_month.dt.to_timestamp()) // pd.Timedelta(days=30)
```

Делаем когортный анализ пользователей

```python
pivot_table = one_year.pivot_table(index='one_month', columns='next_months',values='order_id', aggfunc='count')
for j in range(1, 19):
    pivot_table.iloc[:, j] = pivot_table.iloc[:, j] / pivot_table.iloc[:, 0]
pivot_table[0] = 1
pivot_table
```

```python
cohort_percentage_style = (pivot_table
            .style
            .background_gradient()
            .highlight_null('white')
            .format("{:.2%}", na_rep=""))
cohort_percentage_style
```

```python
# объединение всех файлов
res_rf = ord_cus.merge(order_items, how='inner', on='order_id')
```

```python
# удаляем строки, где есть хотя бы одно NaN 
res_rf = res_rf.dropna(subset=['order_approved_at'])
```

```python
# количество дней между максимальным временем подтверждения оплаты заказа и текущим
res_rf['days_delta'] = (res_rf['order_approved_at'].max() - res_rf['order_approved_at']).dt.days
res_rf
```

```python
"""
recency - время от последней покупки пользователя до текущей даты
frequency - суммарное количество покупок у пользователя за всё время
monetary - сумма покупок за всё время
"""
rs_rf = res_rf.groupby(['customer_unique_id'], as_index=False) \
                                 .agg({'days_delta':'min', 'order_id': 'nunique', 'price': 'sum'}) \
                                 .rename(columns={'days_delta': 'recency'})\
                                 .rename(columns={'order_id':'frequency'})\
                                 .rename(columns={'price': 'monetary'})
rs_rf
```

```python
 def r_score(x):
    if x >= 321:
        return 1
    elif x >= 241:
        return 2
    elif x >= 161:
        return 3
    elif x >= 81:
        return 4
    else:
        return 5
    
rs_rf['r_sc'] = rs_rf['recency'].apply(lambda x: r_score(x))
```

```python
def f_score(y):
    if y == 1:
        return 1
    elif y <= 2:
        return 2
    elif y <= 3:
        return 3
    elif y <= 4:
        return 4
    else:
        return 5
    
rs_rf['f_sc'] = rs_rf['frequency'].apply(lambda y: f_score(y))
```

```python
def m_score(z):
    if z <= 47:
        return 1
    elif z <= 89:
        return 2
    elif z <= 154:
        return 3
    elif z <= 200:
        return 4
    else:
        return 5
    
rs_rf['m_sc'] = rs_rf['monetary'].apply(lambda z: m_score(z))
```

```python
# RFM-сегментация пользователей
rs_rf['RFM_sc'] = rs_rf['r_sc'].map(str) + rs_rf['f_sc'].map(str) + rs_rf['m_sc'].map(str)
```

```python
rs_rf
```

```python
# кластеризация покупателей
segt_map = {
    r'5[3-5][4-5]': 'Champions', #активные клиенты
    r'[3-4][2-5][2-5]|533|532': 'Loyal Customers', #лояльные клиенты
    r'52[1-5]|31[2-5]|41[1-5]|421|321|311': 'Potential Loyalists', #потенциально лояльные клиенты
    r'[1-2][2-4][1-5]|152': 'About To Sleep', #давно не совершали покупки
    r'[1-2]1[1-5]': 'Hibernating', #давно не было покупок, мало заказов
    r'51[1-5]': 'New Customers' #новые клиенты
}

rs_rf['rest'] = rs_rf['r_sc'].astype(str) + rs_rf['f_sc'].astype(str) + rs_rf['m_sc'].astype(str)
rs_rf['rest'] = rs_rf['rest'].replace(segt_map, regex=True)
rs_rf.head()
```

```python
segments_counts = rs_rf['rest'].value_counts().sort_values(ascending=True)

fig, ax = plt.subplots()

bars = ax.barh(range(len(segments_counts)),
              segments_counts,
              color='silver')
ax.set_frame_on(False)
ax.tick_params(left=False,
               bottom=False,
               labelbottom=False)
ax.set_yticks(range(len(segments_counts)))
ax.set_yticklabels(segments_counts.index)

for i, bar in enumerate(bars):
        value = bar.get_width()
        if segments_counts.index[i] in ['Champions', 'Loyal Customers']:
            bar.set_color('firebrick')
        ax.text(value,
                bar.get_y() + bar.get_height()/2,
                '{:,} ({:}%)'.format(int(value),
                                   int(value*100/segments_counts.sum())),
                va='center',
                ha='left'
               )

plt.show()
```

1) В результате проведенного анализа наблюдается больше число пользователей (43%),

2) Среднее суммарное количество покупок для всех пользователей за всё время составило 1.03.

3) Среди покупателей 17% составляют новые клиенты, которые совершили заказ в течение последних 80 дней.
